```python

import json
import random

cpp_json_path = "../RAW_DATA/cpp.json"

with open(cpp_json_path, "r") as f:
    data = json.load(f)

assert len(data) == 4
block_count = len(data[0])

print(block_count)
```

```python
sample_count = 10

for i in range(sample_count):
    # random choose an index from 0 to block_count - 1
    index = random.randint(0, block_count - 1)
    base_content = data[0][index]
    our_content = data[1][index]
    their_content = data[2][index]
    res_content = data[3][index]
    print(f"""
<<<<<<<
{our_content}
=======
{base_content}
|||||||
{their_content}
>>>>>>>
truth:
{res_content}\n\n\n
""")
```

```python
cpp_train_json = "../RAW_DATA/split/train.json"
cpp_val_json = "../RAW_DATA/split/val.json"
cpp_test_json = "../RAW_DATA/split/test.json"

with open(cpp_train_json, "r") as f:
    train_data = json.load(f)

with open(cpp_val_json, "r") as f:
    val_data = json.load(f)

with open(cpp_test_json, "r") as f:
    test_data = json.load(f)

print(f"train: {len(train_data[0])}")
print(f"val: {len(val_data[0])}")
print(f"test: {len(test_data[0])}")

# random choose 5 for display
for i in range(5):
    index = random.randint(0, len(train_data[0]) - 1)
    base_content = train_data[0][index]
    our_content = train_data[1][index]
    theirs_content = train_data[2][index]
    res_content = train_data[3][index]
    print(f"""
<<<<<<<
{our_content}
=======
{base_content}
|||||||
{theirs_content}
>>>>>>>
truth:
{res_content}\n\n\n
""")
```
